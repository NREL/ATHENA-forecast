```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

sns.set()
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("single_values.csv")
df = df.drop("Unnamed: 0", axis=1)
piv = df.pivot_table(index=['day', 'tw', 'hw', 'mr', 'dr'], 
                     columns=['attribute'], 
                     values='value').reset_index()

piv['dr'] = piv['dr'].map(lambda x: round(x,1))
```

```python
sns.lmplot(x="number_of_buses", y="hourly_cost", data=piv, height=6)
plt.grid(False)
```
